## 2nd project

```python
X_New, Y_New = make_classification(n_samples= 1000, n_features=10, n_classes=2, random_state=45)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_New, Y_New, test_size=0.30, random_state=45)
```

```python
x_train.shape, x_test.shape
```

```python
regressor = LogisticRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## Hyperparameter Tuning And Cross Validation

## Grid SearchCV

```python
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
cv=StratifiedKFold()
```

```python
grid=GridSearchCV(estimator=regressor,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_prd = grid.predict(x_test)
```

```python
score=accuracy_score(y_prd,y_test)
print(score)
print(classification_report(y_prd,y_test))
print(confusion_matrix(y_prd,y_test))
```

## Randomized SearchCV

```python
randomcv=RandomizedSearchCV(estimator=regressor,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(x_train, y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_predc = randomcv.predict(x_test)
```

```python
score=accuracy_score(y_predc,y_test)
print(score)
print(classification_report(y_predc,y_test))
print(confusion_matrix(y_predc,y_test))
```

## Logistic Regression For Multiclass Classification Problem

```python
X_n, y_n = make_classification(n_samples=1000, n_features=10,n_informative=3, n_classes=3, random_state=15)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_n,y_n,test_size=0.30,random_state=42)
```

```python
logistic=LogisticRegression(multi_class='ovr')
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
